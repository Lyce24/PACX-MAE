```python
import os

val_views = []
train_views = []

for dir in os.listdir("/users/yliu802/.cache/kagglehub/datasets/ashery/chexpert/versions/1/valid/"):
    # list all subfiles in a directory joined with "/"
    subfiles = os.listdir(os.path.join("/users/yliu802/.cache/kagglehub/datasets/ashery/chexpert/versions/1/valid/", dir))
    for subfolders in subfiles:
        subfolder = os.listdir(os.path.join("/users/yliu802/.cache/kagglehub/datasets/ashery/chexpert/versions/1/valid/", dir, subfolders))
        for file in subfolder:
            val_views.append((os.path.join(dir, subfolders, file)))

for dir in os.listdir("/users/yliu802/.cache/kagglehub/datasets/ashery/chexpert/versions/1/train/"):
    # list all subfiles in a directory joined with "/"
    subfiles = os.listdir(os.path.join("/users/yliu802/.cache/kagglehub/datasets/ashery/chexpert/versions/1/train/", dir))
    for subfolders in subfiles:
        subfolder = os.listdir(os.path.join("/users/yliu802/.cache/kagglehub/datasets/ashery/chexpert/versions/1/train/", dir, subfolders))
        for file in subfolder:
            train_views.append((os.path.join(dir, subfolders, file)))
```

```python
# Convert lists to sets for faster lookup
train_views_set = set(train_views)
val_views_set = set(val_views)

print(len(train_views))
print(len(val_views))

# Filter into train / val subsets
df_train = df[df["Path"].isin(train_views_set)].reset_index(drop=True)
df_val = df[df["Path"].isin(val_views_set)].reset_index(drop=True)

# Optional: check sizes
print("Train samples:", len(df_train))
print("Validation samples:", len(df_val))

# Optional: save to CSV
df_train.to_csv("train_split.csv", index=False)
df_val.to_csv("val_split.csv", index=False)
```
