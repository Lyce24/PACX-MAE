```python
import pandas as pd

# get from https://physionet.org/content/chexchonet/1.0.0/ 
df = pd.read_csv("../src/metadata.csv")
df.head()
```

```python
# only keep Path and Label columns
df = df[["cxr_filename", "composite_slvh_dlv"]]

df.head()
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['composite_slvh_dlv'])
train_df, val_df = train_test_split(train_df, test_size=0.1, random_state=42, stratify=train_df['composite_slvh_dlv'])

# save to csv
train_df.to_csv("../src/chexchonet_train.csv", index=False)
val_df.to_csv("../src/chexchonet_val.csv", index=False)
test_df.to_csv("../src/chexchonet_test.csv", index=False)
```
